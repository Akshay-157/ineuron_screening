2. Demonstrate use of abstract class, multiple inheritance and decorator in
python using examples.

```python
from abc import ABC, abstractmethod
 
class Shape(ABC): #abstract class which will contain the blueprint for specific shapes 
    def __init__(self, shape_name):
        self.shape_name = shape_name
    
    @abstractmethod
    def draw(self):
        pass
```

```python
class Circle(Shape): #inheriting the abstract class
    def __init__(self):
        super().__init__("Triangle")
 
    def draw(self): #implementing the abstract method
        print("Drawing a Triangle")
```

```python
circle = Circle()
circle.draw()
```

```python
# Parent class1
class Parent1:
    def parent1_func(self):
        print("Hi I am first Parent")

# Parent class2
class Parent2:
    def parent2_func(self):
        print("Hi I am second Parent")

# Child class
class Child(Parent1, Parent2): #child class inheriting parent class1 and parent class2
    def child_func(self):
        #the members of both the parent classes in available to the child class
        self.parent1_func()
        self.parent2_func()

#creating object for the child class
obj1 = Child()
obj1.child_func()
```

```python
#decorators
"""
The Decorators in python are essentially functions that add functionality to an existing function in python
without changing the structure of the function itself. They are represented using "@deorator_name" in python
and are called in a bottom-up fashion.
"""
def names(function): #decorator function
    def wrapper(arg1, arg2):
        #inner function adding additonal functionality by capitalizing the parameters
        arg1 = arg1.capitalize()
        arg2 = arg2.capitalize()
        return function(arg1, arg2) #return the function with updated arguments.
    return wrapper #return decorator

@names #applying the decorator on the function hello.
def hello(name1, name2):
    return "Hello " + name1 + "! Hello " + name2 + "!"
```

```python
hello("hulk", "spidy")
#in the output, we can observe that the parameters are capitalized.
```
